```python
from selenium.common.exceptions import  NoSuchElementException, ElementClickInterceptedException
import time
driver = Chrome()
driver.get('https://maps.foundationcenter.org/#/list/?subjects=all&popgroups=all&years=2020&location=6254926&excludeLocation=0&geoScale=ADM1&layer=geo_area&boundingBox=-138.5595703125,30.977609093348686,-74.2236328125,50.680797145321655&gmOrgs=all&recipOrgs=all&tags=all&keywords=&pathwaysOrg=&pathwaysType=&acct=raceequpopup&typesOfSupport=all&transactionTypes=all&amtRanges=all&minGrantAmt=0&maxGrantAmt=0&gmTypes=all&recipTypes=all&minAssetsAmt=0&maxAssetsAmt=0&minGivingAmt=0&maxGivingAmt=0&andOr=0&includeGov=1&custom=all&customArea=all&indicator=&dataSource=oecd&chartType=trends&multiSubject=1&listType=gm&windRoseAnd=undefined&zoom=5')
time.sleep(2)

# Closes the pop-up window
close_button = driver.find_element(By.CSS_SELECTOR, 'button[class="close"]')
close_button.click()


current_page = 1

while True:
    try:
        # Wait for the pagination elements to load properly
        time.sleep(2)  # Adjust time as needed based on the responsiveness of the website

        # Attempt to find and click the next page number
        next_page = current_page + 1
        next_page_element = driver.find_element(By.LINK_TEXT, f'{next_page}')
        next_page_element.click()
        # print(next_page_element.text)

        print(f"Moved to page {next_page}")
        
        # Update current_page to the new page number after successful click
        current_page = next_page

    except NoSuchElementException:
        # If no more pages can be found to click, break the loop
        print("No more pages to load.")
        break
    except ElementClickInterceptedException:
        # Handle cases where the click is blocked by other elements, e.g., overlays
        print("Click was blocked by another element, trying again.")
        time.sleep(1)  # Waiting a bit before trying again

print("Done processing all pages.")
```
